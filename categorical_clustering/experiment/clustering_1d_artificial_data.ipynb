```python
import pandas as pd

from categorical_clustering.experiment.label_generator import generate_labels
labels = generate_labels({'A': 20, 'B': 10, 'C': 5, 'D': 15})
df = pd.DataFrame(data={'Labels': labels})
```

```python
from categorical_clustering.core.algorithm.separate_attribute_pmf import SeparateAttributePMF as SAPMF
model = SAPMF(n_clusters=3)
model.set_data(df)
```

```python
clustering, iters = model.perform_clustering(update_proportion_criterion=1e-2, verbose=True)
print('Converged in {} iterations!'.format(iters))
print('Clusters:')
model.clustering.clusters
```

```python
import chart_studio.plotly as py
import plotly.graph_objects as go

df = model.get_cluster_as_dataframe(cluster=-1)

fig = go.Figure(data=[
    go.Bar(x=df.index, y=df.iloc[:, 0])
])
fig.update_xaxes(title='Labels')
fig.update_yaxes(title='Counts')
fig.update_layout(title='Overall Data', autosize=False, width=500, height=400)
fig.show()
```
